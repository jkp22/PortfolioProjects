# Visualizations

```python
import matplotlib.pyplot as plt

%matplotlib inline 
plt.style.use('seaborn')
import seaborn as sns
```

### Total steps vs Cals

```python
sns.regplot(x='TotalSteps',y='Calories', data = activity)
font1 = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 16,
        }
plt.title('Total Steps vs Calories', fontdict = font1)
plt.xlabel('Totel steps', fontdict = font1)
plt.ylabel('Calories',fontdict = font1)
```

#### Nothing suprising here. The more steps you do the more cals you burn

## Looking at intentsity grouped by time

```python
inten2 = inten.groupby('Time').agg({
    'TotalIntensity' : 'mean',
    
}).dropna()
inten2 = inten2.reset_index()
inten2.head()
```

```python
plt.figure(figsize=(8,4))
int_grpah = sns.barplot(data=inten2,x='Time',y='TotalIntensity')
int_grpah.set_xticklabels(int_grpah.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.title('Avg. Intensity by time of day', fontdict = font1)
plt.xlabel('Time', fontdict = font1)
plt.ylabel('Average Intensity',fontdict = font1)
plt.show()
```

*Here we see that that we see the most time spent under intense load is the times between 5 and 7pm. Most likly due to going after work*
